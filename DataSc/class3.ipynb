```python
import pandas as pd
```

```python
car_sales=pd.read_csv("car-sales.csv")
```

```python
car_sales
```

## Creating a series 

```python
seats=pd.Series([4,4,4,])
```

```python
seats
```

## Adding a series as column to an exisiting data frame

```python
car_sales["seats"]=seats
car_sales
```

## Filling the NAN with valu '4' in the data frame

```python
car_sales["seats"].fillna(4, inplace=True)
car_sales
```

```python
car_sales.head(4)
```

## adding a list to a newly assigned column

```python
fuel_economy = [1,8,9,2,4,5,6,4,2,0]
car_sales["Fuel per 100KM"]=fuel_economy
car_sales
```

## Adding an arithmetic operator to a newly assigned column

```python
car_sales["Total fuel used"]= car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
```

```python
car_sales
```

## Assigning a value to a new column

```python
car_sales["Number of wheels"]=4
car_sales.head(5)
```

## Adding a boolan column to a data frame

```python
car_sales["passed road saftey"]= True
car_sales
```

## Dropping a column

```python
car_sales= car_sales.drop("Total fuel used", axis=1)
car_sales


```

```python
car_sales= car_sales.drop("index", axis=1)
car_sales
```

```python
car_sales= car_sales.drop("level_0", axis=1)
car_sales
```
